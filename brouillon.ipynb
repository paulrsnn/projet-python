## base qui regroupe les données par sexe

```python
def donnees_par_sexe(base_initial):
    
    base=base_initial[['Code Commune', 'Commune', 'Département','Région', 'Code', 'Fédération']]
    base['F']=base_initial[['F - 5 à 9 ans', 'F - 10 à 14 ans', 'F - 15 à 19 ans',
              'F - 20 à 24 ans', 'F - 25 à 29 ans', 'F - 30 à 34 ans',
              'F - 35 à 39 ans', 'F - 40 à 44 ans', 'F - 45 à 49 ans',
              'F - 50 à 54 ans', 'F - 55 à 59 ans', 'F - 60 à 64 ans',
              'F - 65 à 69 ans', 'F - 70 à 74 ans', 'F - 75 à 79 ans',
              'F - 80 à 99 ans', 'F - NR']].sum(axis=1)
    base['H']=base_initial[['H - 1 à 4 ans', 'H - 5 à 9 ans',
              'H - 10 à 14 ans', 'H - 15 à 19 ans', 'H - 20 à 24 ans',
              'H - 25 à 29 ans', 'H - 30 à 34 ans', 'H - 35 à 39 ans',
              'H - 40 à 44 ans', 'H - 45 à 49 ans', 'H - 50 à 54 ans',
              'H - 55 à 59 ans', 'H - 60 à 64 ans', 'H - 65 à 69 ans',
              'H - 70 à 74 ans', 'H - 75 à 79 ans', 'H - 80 à 99 ans', 'H - NR']].sum(axis=1)
    base['NR']=base_initial[['NR - 5 à 9 ans', 'NR - 10 à 14 ans', 'NR - 15 à 19 ans',
                 'NR - 30 à 34 ans', 'NR - 40 à 44 ans', 'NR - 45 à 49 ans',
                 'NR - 70 à 74 ans', 'NR - NR']].sum(axis=1)
    return base
```

```python
base_par_sexe= donnees_par_sexe(lic_comm_sup)
base_par_sexe
```

```python


def lic_reg_per_sex(base_initial):
    base=base_initial[base_initial.columns.difference(['Code Commune', 'Commune','Département','Code', 'Fédération'])]
    base1=base.groupby('Région').sum().reset_index()
    base1['Total_m']=base1['H']+base1['F']+base1['NR']
    base1['taux_F']=(base1['F']/base1['Total_m'])* 100
    base1['taux_H']=(base1['H']/base1['Total_m'])* 100
    base1['taux_NR']=(base1['NR']/base1['Total_m'])* 100
    base1=base1.drop(['F','H','NR','Total_m'],axis=1)
    return base1
```

```python
def sport_plus_pratique_par_dep(base_initial):
    base=base_initial[base_initial.columns.difference(['Code Commune', 'Commune','Région'])]
    base1=base.groupby(['Département','nom_departement','Code', 'Fédération']).sum().reset_index()
    base1['nbre_licences']=base1[['H','F','NR']].sum(axis=1)
    base1=base1.drop(['F','H','NR'],axis=1)
    #labels_lignes_max=base1.groupby('Département')['nbre_licences'].idxmax()
   # ligne_max=base1.loc[labels_lignes_max][['Département','nom_departement','Code', 'Fédération','nbre_licences']]
    return base1
```
